```python
# load required pkg

import tkinter as tk
from tkinter import filedialog, ttk
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')



# create a TKinter root window
#root = tk.Tk()
# hide the root window
#root.withdraw()

# create a file dialog object
#file_path = filedialog.askopenfilename(title='Select a File', filetypes=(("All files", "*.csv"),))

#print("Selected file path: ", file_path)

# dataset path
file_path=('winequality_red.csv')
dataset = pd.read_csv(file_path, sep=';')
pd.set_option('display.max_columns', None)



class Data_preparation():
    global dataset
    class EDA():
        # create a functio to view dataset
        def dataset_head(self,data_set):
            print('=============================================')
            print('  Let\'s check dataset columns and data')
            print('=============================================')
            display(data_set.head())
            
        # create a function to seperate the text and numeric columns
        def sep_data_types(self,data_set):
            print('=============================================')
            print('  Seperate Numeric and Text columns ')
            print('=============================================')
            num_col = data_set.select_dtypes(include = ['int64', 'float64']).columns.tolist()
            txt_col = data_set.select_dtypes(include = ['object']).columns.tolist()
            print('=============================================')
            print('   Numeric Columns:')
            print('=============================================')
            for col in num_col:
                print(col)
            print('=============================================')
            print('  Text Columns:')
            print('=============================================')
            for col in txt_col:
                print(col)
                
        # create a function to find the unique characters of each columns
        def unq_char(self, data_set):
            print('=============================================')
            print('  Unique Characters in Each Column')
            print('=============================================')
            for i in data_set.columns:
                unq = data_set[i].unique()
                print(f"column'{i}' has {len(unq)} unique values:\n{unq}\n")
                
        # create a function to view the shape of the dataset
        def dataset_shape(self, data_set):
            print('=============================================')
            print('  Shape of the Dataset')
            print('=============================================')
            print(data_set.shape)
            
        # create a function to view the dataset information
        def dataset_info(self, data_set):
            print('=============================================')
            print('  Dataset Info')
            print('=============================================')
            print(data_set.info())
            
        # create a function to view the descriptive statistics of the dataset
        def des_sta(self, data_set):
            print('=============================================')
            print('  Descriptive Statistics')
            print('=============================================')
            display(data_set.describe())
            
        # create a function to view the outliers
        def out_det(self, data_set):
            print('=============================================')
            print('  Outliers Visualization')
            print('=============================================')
            fig, axs = plt.subplots(nrows=1, ncols=len(data_set.columns), figsize=(3*(len(data_set.columns)),len(data_set.columns)))
            for i, col in enumerate(data_set.columns):
                axs[i].boxplot(data_set[col])
                axs[i].set_title(col)
                
            plt.show()
            
        # create a function to view the number of duplicate values
        def dup_val(self, data_set):
            print('=============================================')
            print('  Duplicate Values')
            print('=============================================')
            print('\n')
            print(f'The dataset contains duplicate rows:', data_set.duplicated().sum())
            
        # create a function to view the null values
        def null_val(self, data_set):
            print('=============================================')
            print('  Null Values')
            print('=============================================')
            null_mask = data_set.isnull()
            null_mask.sum().plot(kind='bar')
            plt.title('Null values by column')
            plt.xlabel('columns')
            plt.ylabel('Number of Null Values')
            plt.show()
            
data_pre = Data_preparation.EDA()
#data_pre.dataset_head(dataset)
#data_pre.sep_data_types(dataset)
#data_pre.unq_char(dataset)
#data_pre.dataset_shape(dataset)
#data_pre.dataset_info(dataset)
#data_pre.des_sta(dataset)
#data_pre.out_det(dataset)
#data_pre.dup_val(dataset)
#data_pre.null_val(dataset)


print('==================================================')
print('        Automate The Data Preparation')
print('==================================================')

opt_list = ['Let\'s check dataset columns and data', 'Seperate Numeric and Text columns', 'Unique Characters in Each Column', 'Shape of the Dataset', 'Dataset Information', 'Descriptive Statistics', 'Outliers Visualization', 'Duplicate Values', 'Null Values']

for i, item in enumerate(opt_list):
    print(f"{i+1}- {item}")
print("Enter 0 to view all the available methods:")

in_put = input('Enter your option as number:')
try:
    num_int = int(in_put)
    if num_int == 1:
        data_pre.dataset_head(dataset)
    elif num_int == 2:
        data_pre.sep_data_types(dataset)
    elif num_int == 3:
        data_pre.unq_char(dataset)
    elif num_int == 4:
        data_pre.dataset_shape(dataset)
    elif num_int == 5:
        data_pre.dataset_info(dataset)
    elif num_int == 6:
        data_pre.des_sta(dataset)
    elif num_int == 7:
        data_pre.out_det(dataset)
    elif num_int == 8:
        data_pre.dup_val(dataset)
    elif num_int == 9:
        data_pre.null_val(dataset)
    elif num_int == 0:
        data_pre.dataset_head(dataset)
        data_pre.sep_data_types(dataset)
        data_pre.unq_char(dataset)
        data_pre.dataset_shape(dataset)
        data_pre.dataset_info(dataset)
        data_pre.des_sta(dataset)
        data_pre.out_det(dataset)
        data_pre.dup_val(dataset)
        data_pre.null_val(dataset)
        
    
except ValueError:
    print("Enter only Integer Value as shown")
```
